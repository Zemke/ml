```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
im = cv2.imread("handwritten.jpg")
im = cv2.bitwise_not(im)  # invert
im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
im = im / 255 * .99 + .01
im = np.minimum(im * 3.5, 1)
im = np.clip(im, .1, 1)
```

```python
plt.imshow(im)
plt.show()
```

```python
def raster(coo, w=40):
    return coo + w

coords = [(272, 161), (48, 30), (47, 113), (45, 181), (45, 245),
          (48, 303), (155, 35), (147, 110), (149, 209), (140, 292)]
sep = []
for c in range(len(coords)):
    y,x = coords[c]
    sep.append(cv2.resize(im[y:raster(y), x:raster(x)], dsize=(28, 28)))
    
subplots = plt.subplots(2, 5)
si = 0
for spa in subplots[1]:
    for sp in spa:
        sp.imshow(sep[si])
        si += 1
plt.show()
```
