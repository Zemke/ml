# Numerical Differentiation

Calculating approximations of derivatives using two points that are "sufficiently" close rather than infinitely close to understand the concepts of derivatives and tangent lines of a non-linear function.

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
def f(x):
    return 2*x**2
````

````python
x = np.arange(0, 5, .1)
y = f(x)  # talking numpy: y for every x (element-wise)
plt.plot(x, y)
````

````python
p2_delta = .000001  # the "sufficiently" close other point
x1 = 2
x2 = x1 + p2_delta
y1 = f(x1)
y2 = f(x2)
x1,x2,y1,y2
````

Remember a linear function is `y = mx + b`. The tangent line is a linear function. `m` is the approximate derivative and `x` is the function input which leaves `b` for calculating. Shifting the term gets us `b = y - mx`.
The tangent line is the slope at a point in a curve. In other words the rate of change.

The formular for getting the slope is `∆y / ∆x`.

````python
approx_derivative = (y2-y1) / (x2-x1)
b = y2 - approx_derivative * x2
b, approx_derivative
````

````python
def tangent_line(m, x, b):
    return m * x + b
````

````python
plt.plot(x, y)
xx = [x1-.9, x1, x1+.9]
yy = [tangent_line(x=x, m=approx_derivative, b=b) for x in xx]
plt.plot(xx, yy)
plt.show()
x1, approx_derivative
````

Many tangent lines displaying the slopes at different points.

````python
for i in range(0, 50, 10):
    xx = [x[i]-.9, x[i], x[i]+.9]
    approx_derivative = (f(x[i]+p2_delta) - f(x[i])) / (x[i]+p2_delta - x[i])
    b = y[i] - approx_derivative * x[i]
    yy = [tangent_line(x=x, m=approx_derivative, b=b) for x in xx]
    plt.plot(xx, yy)
    plt.scatter(xx[1], yy[1])
plt.plot(x, y)
plt.show()
````

For a linear functions this is more simple as the two points don't need to be close because the linear slope is always the same.

```
ƒ(x) = mx + b = y
```

In the following case for every change in `x`, `y` is 2 times `x` which is the slope.

````python
# def linear_f(x, b=0, m=2):
#     return m * x + b  # m is the slope

xx = np.arange(0, 6)
yy = [y for y in tangent_line(x=xx, b=0, m=2)]
plt.plot(xx, yy)

slope = (yy[-1] - yy[0]) / (xx[-1] - xx[0])  # ∆y / ∆x
xx_plot = xx[2:4]
plt.plot(xx_plot, tangent_line(x=xx_plot, b=-.2, m=slope))  # slightly offset by b to make it more visible in the graph
slope, xx, yy
````

> The **derivative** is the **slope** of the **tangent line**.

For linear functions the slope equals the derivative.
